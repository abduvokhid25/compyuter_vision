```python
import matplotlib.pyplot as plt
import numpy as np
```

Basic LIne PLot

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])

plt.plot(x, y)
plt.title('Basic Line Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([3, 8, 1, 10])

plt.plot(y, linestyle='dotted', color='green', marker='o')
plt.title('Styled Line Example')
plt.show()
```

Multiople Line one Graph

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([1, 3, 5, 7])

plt.plot(x, y1, label='line1', color='red', marker='s')
plt.plot(x, y2, label='line2',  color='blue', linestyle='--', marker='o')
plt.title('Multiple Line Examples')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.show()
```

Barchart with numpy and MatPlotLip

```python
x = np.array(['Salary', 'Expences', 'Charity', 'Food'])
y = np.array([5, 9, 1, 7])

plt.bar(x, y, color='orange')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

Scatter with numpy and MatPlotLip

```python
x = np.array([1, 2, 2.5, 3, 4])
y = np.array([3, 5, 3.5, 1, 7])

plt.scatter(x, y, color='purple', s=200)
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

Histogram with Numpy and MatPlotLib

```python
data = np.array([3, 5, 6, 2, 8, 10])

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Count')
plt.show()
```

SubPlots(Multiple Graph togenther)

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([3, 6, 9, 12])

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red')
plt.title('Plot1')

plt.subplot(1, 2, 2)
plt.bar(x, y2, color='blue')
plt.title('Plot 2')
plt.tight_layout()
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([4, 8, 6, 10])

plt.plot(x, y, color='green')
plt.title('Save Figure Example')
plt.savefig('my_plot.png') # Saves in current folder if you are working in vscode
plt.show()
```

```python
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import axes3d

plt.style.use('_mpl-gallery')

# Make data
X, Y, Z = axes3d.get_test_data(0.05)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
np.random.seed(19680801)
n = 100
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y1 = np.array([10, 15, 20, 25])
y2 = np.array([8, 12, 18, 22])
plt.plot(x, y1, label='Dataset A', color='red', marker='o')
plt.plot(x, y2, label='Dataset B', color='blue', marker='s', linestyle='--')
plt.title('Matplotlib with Numpy Arrays')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid(True)
plt.show()
```
